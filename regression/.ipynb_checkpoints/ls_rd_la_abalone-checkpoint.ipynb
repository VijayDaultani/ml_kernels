# Step 1 : Import Necessary Libraries

```python
"""
First of all we will import necessary packages required for our tasks of exploration,
visualization, analysis and fitting models and measuring how good they are
"""
#Import numpy and pandas
import numpy as np
import pandas as pd

#Import seaborn and matplotlib for plotting
import seaborn as sns
import matplotlib.pyplot as plt

#Below line is used to make sure matplotlib plots all graphs on the same notebook
%matplotlib inline
```

# Step 2 : Load data 

```python
"""
Load data in abalone_df Pandas DataFrame
We use delimeter= " " because in the dataset the columns are separated using " "
Also we specify names of the column explicitly to ease our understanding.
"""
abalone_df = pd.read_csv("../../data/abalone/Dataset.data", delimiter=" ",
                         names=['Sex', 'Length', 'Diameter', 'Height', 
                                'Whole weight', 'Shucked weight', 'Viscera weight',
                                'Shell weight', 'Rings'])
```

# Step 3 : Start exploring dataset 

```python
#First of all lets look how big the dataset is using len function
print "Number of Observations in the dataset : " + str(len(abalone_df))

#Now lets have a glimpse of first few rows of the dataset
abalone_df.head()
```

```python
#Now lets have a glimpse of last few rows of the dataset
abalone_df.tail()
```

```python
#Now lets print the shape of the Dataframe (rows, columns)
abalone_df.shape
```

# Step 4 : Lets use seaborn library to visualize the dataset

```python
sns.pairplot(abalone_df, x_vars=['Length','Diameter','Height','Whole weight','Shucked weight'
                                 ,'Viscera weight','Shell weight'], y_vars='Rings',hue='Sex'
                                 , size=4, aspect=0.7)
```

```python
#Lets try to how does the relationship goes between features and Rings
sns.pairplot(abalone_df, x_vars=['Length','Diameter','Height','Whole weight','Shucked weight'
                                 ,'Viscera weight','Shell weight'], y_vars='Rings'
                                 , size=4, aspect=0.7, kind='reg')
```

# Step 5 : Preparing X and y using pandas

```python
#First of all we should create a list of feature names
features = ['Length','Diameter','Height','Whole weight','Shucked weight','Viscera weight',
            'Shell weight']

#Use these list as to select the subset of Series from pandas dataframe
X = abalone_df[features]

#Lets have a look at how X looks
print X.head()
```

```python
#Also lets check the type of X and shape of X
print type(X)
print X.shape
```

```python
#Select output or target Series which we want to predict
y = abalone_df['Rings']

y.head()
```

```python
#Check the type and shape of y
print type(y)
print y.shape
```

# Step 6 : Splitting X and y into training and testing sets

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)

#Since scikit does 75% for training and 25% for testing by default
print X_train.shape
print X_test.shape
print y_train.shape
print y_test.shape
```

# Step 7 : Linear regression using scikit-learn

```python
#import model
from sklearn.linear_model import LinearRegression

#Create a object of Linear Regression model
linreg = LinearRegression()

#Now perform the fitting of model on data
linreg.fit(X_train,y_train)
```

```python
#Printing the coefficients learned
print linreg.intercept_
print linreg.coef_
```

```python
#Or also we can print them together with name of features
zip(features, linreg.coef_)
```

```python
#Plotting the output
plt.scatter(X_test.Length, y_test,color='black')
plt.plot(X_test.Length, linreg.predict(X_test), color='blue')
#print X_train.shape
#print y_train.shape
```

# Step 8 : Make predictions using model learned

```python
#We will make predictions using the testing set
y_pred = linreg.predict(X_test)
```

# Step 9 : Evaluate model using RMSE

```python
from sklearn import metrics
print np.sqrt(metrics.mean_squared_error(y_test,y_pred))
```

# Step 10 : Perform the cross validation

```python
from sklearn.cross_validation import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
print X_train.shape
print X_train.head()
print y_train.shape
print y_train.head()
```

```python
#10-fold cross validation with RMSE
# LinearRegression(fit_intercept=True, normalize=False, copy_X=True, n_jobs=1)
regr = LinearRegression()
scores = cross_val_score(regr, X_train, y_train, cv=10, scoring='mean_squared_error')
print scores
```
