# Setup
## Step 1 : Import Necessary Libraries

```python
"""
First of all we will import necessary packages required for our task of exploration,
visualization, analaysis and fitting models and measuring how good they are
"""

#Import numpy and pandas
import numpy as np
import pandas as pd

#Import seaborn and matplotlib for plotting
import matplotlib.pyplot as plt 
import seaborn as sns

#Magic line to print all matplotlib plots on the same notebook
%matplotlib inline
```

# Step 2 : Load Data

```python
"""
Load data in abalone_df Pandas DataFrame
We use delimeter= " " because in the dataset the columns are separated using " "
Also we specify names of the column explicitly to ease our understanding.
"""
cpu_df = pd.read_csv("../../../data/cpu_performance/machine.data",
                    names=['Vendor','Model','MYCT', 'MMIN', 'MMAX', 'CACH', 'CHMIN', 'CHMAX',
                          'PRP', 'ERP'])
```

# Step 3 : Start exploring dataset 

```python
#Now lets print the shape of the Dataframe (rows, columns)
print cpu_df.shape
#Now lets have a glimpse of first few rows of the dataset
cpu_df.head()
```

# Step 4 : Lets use seaborn library to visualize the dataset

```python
"""
Let's try to visualize how the dataset looks like. Since our traget is Rings we will use
seaborns pairplot to visualize the relationship between different various features or 
variables with respect to target i.e. Rings

Such kind of analysis will help us to know for Ordinary least squares which are the variables
which we should be using for regression.
"""
sns.pairplot(cpu_df, x_vars=['MYCT', 'MMIN', 'MMAX', 'CACH', 'CHMIN','CHMAX'],
                     y_vars='PRP', hue='Vendor', size=4, aspect=0.7)
```

```python
"""
To make it more easy for the visualization of the data I will fit a linear regression model
using the built in feature of seaborn. 

The below graphs show the if we fit a linear regresion model to each of the feature seprately 
than what would be the line of fit.

Lets try to how does the relationship goes between features and PRP
"""
sns.pairplot(cpu_df, x_vars=['MYCT', 'MMIN', 'MMAX', 'CACH', 'CHMIN','CHMAX'],
                     y_vars='PRP', size=4, aspect=0.7, kind='reg')
```

# Step 5 : Preparing X and y

```python
#First of all we should create a list of feature names
features = ['MYCT', 'MMIN', 'MMAX', 'CACH', 'CHMIN','CHMAX']

#Use these list as to select the subset of Series from pandas dataframe
X = cpu_df[features]
y = cpu_df['PRP']

#Print the dimensions of X and y
print "Dimensions of X : " + str(X.shape)
print "Dimensions of y : " + str(y.shape)
```

# Step 6 : Splitting X and y into training and testing sets

```python
from sklearn.cross_validation import train_test_split
#By default sklearn divides 75% training data and 25% testing data
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=0)
print "Dimensions of X_train : " + str(X_train.shape)
print "Dimensions of X_test : "  + str(X_test.shape)
print "Dimensions of y_train : " + str(y_train.shape)
print "Dimensions of y_test : "  + str(y_test.shape)
```

# Step 7 : Linear regression using scikit-learn

```python
from sklearn.linear_model import LinearRegression

#Create a object of the LinearRegression
lin_reg = LinearRegression()

#Now perform the fitting of model on data
lin_reg.fit(X_train,y_train)
```

```python
#Print intercept and coefficients
print "'Intercpt', " + str(lin_reg.intercept_)
zip(features, lin_reg.coef_)
```

# Step 8 : Predictions on testing data and metrics

```python
"""
Since we cannot draw 3 dimensions on X axis. Right now I am just calculating the metrics
to measure the performance of the model learned.
"""
#Make predictions on dataset
y_pred_lin = lin_reg.predict(X_test)

#The mean RSS 
RSS_lin = np.mean((y_pred_lin - y_test)**2)
print "Linear Regression RSS : " + str(RSS_lin)

#The RMSE
RMSE_lin = np.sqrt(RSS_lin)
print "Linear Regression RMSE : " + str(RMSE_lin)
```

```python
# Variance score : 1 is perfect prediction
print "Variance score : " + str(lin_reg.score(X_test, y_test))
```
