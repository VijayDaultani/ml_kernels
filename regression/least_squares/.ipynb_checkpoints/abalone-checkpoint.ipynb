# Setup
## Step 1 : Import Necessary Libraries

```python
"""
First of all we will import necessary packages required for our task of exploration,
visualization, analaysis and fitting models and measuring how good they are
"""

#Import numpy and pandas
import numpy as np
import pandas as pd

#Import seaborn and matplotlib for plotting
import matplotlib.pyplot as plt 
import seaborn as sns

#Magic line to print all matplotlib plots on the same notebook
%matplotlib inline
```

# Step 2 : Load Data

```python
"""
Load data in abalone_df Pandas DataFrame
We use delimeter= " " because in the dataset the columns are separated using " "
Also we specify names of the column explicitly to ease our understanding.
"""
abalone_df = pd.read_csv("../../../data/abalone/Dataset.data", delimiter=" ",
                         names=['Sex', 'Length', 'Diameter', 'Height', 
                                'Whole weight', 'Shucked weight', 'Viscera weight',
                                'Shell weight', 'Rings'])
```

# Step 3 : Start exploring dataset 

```python
#Now lets print the shape of the Dataframe (rows, columns)
print abalone_df.shape
#Now lets have a glimpse of first few rows of the dataset
abalone_df.head()
```

# Step 4 : Lets use seaborn library to visualize the dataset

```python
"""
Let's try to visualize how the dataset looks like. Since our traget is Rings we will use
seaborns pairplot to visualize the relationship between different various features or 
variables with respect to target i.e. Rings

Such kind of analysis will help us to know for Ordinary least squares which are the variables
which we should be using for regression.

From the below graphs one can visualize that feature Length and Diameter seems to carry
redundant information. Also features Whole weight, Shucked weight, Viscera weight and 
Shell weight seems to carry redundant information.
"""
sns.pairplot(abalone_df, x_vars=['Length','Diameter','Height','Whole weight','Shucked weight'
                                 ,'Viscera weight','Shell weight'], y_vars='Rings',hue='Sex'
                                 , size=4, aspect=0.7)
```

```python
"""
To make it more easy for the visualization of the data I will fit a linear regression model
using the built in feature of seaborn. 
The below graphs show the if we fit a linear regresion model to each of the feature seprately 
than what would be the line of fit.

Also from the below graph one can easily look for the high leverage data points in the graph
of Height as you can see two outliers influence the fitted line a lot. Hence we will try to 
remove such data points from our dataset to make our model more roubust.

Lets try to how does the relationship goes between features and Rings
"""
sns.pairplot(abalone_df, x_vars=['Length','Diameter','Height','Whole weight','Shucked weight'
                                 ,'Viscera weight','Shell weight'], y_vars='Rings'
                                 , size=4, aspect=0.7, kind='reg')
```

# Step 5 : Preparing X and y

```python
#First of all we should create a list of feature names
features = ['Length','Height','Whole weight']

#Use these list as to select the subset of Series from pandas dataframe
X = abalone_df[features]
y = abalone_df['Rings']

#Print the dimensions of X and y
print "Dimensions of X : " + str(X.shape)
print "Dimensions of y : " + str(y.shape)
```

# Step 6 : Splitting X and y into training and testing sets

```python
from sklearn.cross_validation import train_test_split
#By default sklearn divides 75% training data and 25% testing data
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=0)
print "Dimensions of X_train : " + str(X_train.shape)
print "Dimensions of X_test : "  + str(X_test.shape)
print "Dimensions of y_train : " + str(y_train.shape)
print "Dimensions of y_test : "  + str(y_test.shape)
```

# Step 7 : Linear regression using scikit-learn

```python
from sklearn.linear_model import LinearRegression

#Create a object of the LinearRegression
lin_reg = LinearRegression()

#Now perform the fitting of model on data
lin_reg.fit(X_train,y_train)
```

```python
#Print intercept and coefficients
print "'Intercpt', " + str(lin_reg.intercept_)
zip(features, lin_reg.coef_)
```

# Step 8 : Predictions on testing data and metrics

```python
"""
Since we cannot draw 3 dimensions on X axis. Right now I am just calculating the metrics
to measure the performance of the model learned.
"""
#Make predictions on dataset
y_pred_lin = lin_reg.predict(X_test)

#The mean RSS 
RSS_lin = np.mean((y_pred_lin - y_test)**2)
print "Linear Regression RSS : " + str(RSS_lin)

#The RMSE
RMSE_lin = np.sqrt(RSS_lin)
print "Linear Regression RMSE : " + str(RMSE_lin)
```

```python
# Variance score : 1 is perfect prediction
print "Variance score : " + str(lin_reg.score(X_test, y_test))
```

# Step 9 : Let's try to see if removing high leverage points help to increase the performance of the model with 3 features

```python
"""
We will verify if two high leverage points are influential data points. We can verify this by 
analyzing if the coefficients or intercepts change by a very large amount if we remove the
high leverage points from the dataset and rerun the model fitting. Well in this case they are
influential observations because removing them from the dataset increase the coefficient for
the Height from 17 to 47 whereas not a very big differnce on other 2 parameters.
"""
#Remove the high leverage points
abalone_df_wo_hl = abalone_df.loc[abalone_df['Height'] <= 0.4]

#First of all we should create a list of feature names
three_features = ['Length','Height','Whole weight']

#Use these list as to select the subset of Series from pandas dataframe
X_three = abalone_df_wo_hl[three_features]
y = abalone_df_wo_hl['Rings']

#By default sklearn divides 75% training data and 25% testing data
X_three_train, X_three_test, y_train, y_test = train_test_split(X_three, y, random_state=0)

#Create a object of the LinearRegression
lin_reg_three = LinearRegression()

#Now perform the fitting of model on data
lin_reg_three.fit(X_three_train,y_train)

#Print intercept and coefficients
print "'Intercpt', " + str(lin_reg_three.intercept_)
print zip(three_features, lin_reg_three.coef_)

#Make predictions on dataset
y_pred_lin_three = lin_reg_three.predict(X_three_test)

#The mean RSS 
RSS_lin_three = np.mean((y_pred_lin_three - y_test)**2)
print "Linear Regression RSS : " + str(RSS_lin_three)

#The RMSE
RMSE_lin_three = np.sqrt(RSS_lin_three)
print "Linear Regression RMSE : " + str(RMSE_lin_three)

# Variance score : 1 is perfect prediction
print "Variance score : " + str(lin_reg_three.score(X_three_test, y_test))
```

# Step 10 : Let's try to see if considering all the parameters would have increased the performance?

```python
"""
As can be seen that even including all the features in our model RMSE doesnot change much
But accuracy changes from 39% to 53%.
"""
#First of all we should create a list of feature names
all_features = ['Length','Diameter','Height','Whole weight','Shucked weight','Viscera weight',
            'Shell weight']

#Use these list as to select the subset of Series from pandas dataframe
X_all = abalone_df[all_features]
y = abalone_df['Rings']

#By default sklearn divides 75% training data and 25% testing data
X_all_train, X_all_test, y_train, y_test = train_test_split(X_all, y, random_state=0)

#Create a object of the LinearRegression
lin_reg_all = LinearRegression()

#Now perform the fitting of model on data
lin_reg_all.fit(X_all_train,y_train)

#Print intercept and coefficients
print "'Intercpt', " + str(lin_reg_all.intercept_)
print zip(all_features, lin_reg_all.coef_)

#Make predictions on dataset
y_pred_lin_all = lin_reg_all.predict(X_all_test)

#The mean RSS 
RSS_lin_all = np.mean((y_pred_lin_all - y_test)**2)
print "Linear Regression RSS : " + str(RSS_lin_all)

#The RMSE
RMSE_lin_all = np.sqrt(RSS_lin_all)
print "Linear Regression RMSE : " + str(RMSE_lin_all)

# Variance score : 1 is perfect prediction
print "Variance score : " + str(lin_reg_all.score(X_all_test, y_test))
```
