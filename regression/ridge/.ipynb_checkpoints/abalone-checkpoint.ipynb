# Setup
## Step 1 : Import Necessary Libraries

```python
"""
First of all we will import necessary packages required for our task of exploration,
visualization, analaysis and fitting models and measuring how good they are
"""

#Import numpy and pandas
import numpy as np
import pandas as pd

#Import seaborn and matplotlib for plotting
import matplotlib.pyplot as plt 
import seaborn as sns

#Magic line to print all matplotlib plots on the same notebook
%matplotlib inline
```

# Step 2 : Load Data

```python
"""
Load data in abalone_df Pandas DataFrame
We use delimeter= " " because in the dataset the columns are separated using " "
Also we specify names of the column explicitly to ease our understanding.
"""
abalone_df = pd.read_csv("../../../data/abalone/Dataset.data", delimiter=" ",
                         names=['Sex', 'Length', 'Diameter', 'Height', 
                                'Whole weight', 'Shucked weight', 'Viscera weight',
                                'Shell weight', 'Rings'])
```

# Step 3 : Start exploring dataset

```python
#Now lets print the shape of the Dataframe (rows, columns)
print abalone_df.shape
#Now lets have a glimpse of first few rows of the dataset
abalone_df.head()
```

# Step 4 : Lets use seaborn library to visualize the dataset

```python
"""
Let's try to visualize how the dataset looks like. Since our traget is Rings we will use
seaborns pairplot to visualize the relationship between various features or  variables 
with respect to target i.e. Rings

From the below graphs one can visualize that feature Length and Diameter seems to carry
redundant information. Also features Whole weight, Shucked weight, Viscera weight and 
Shell weight seems to carry redundant information.
"""
sns.pairplot(abalone_df, x_vars=['Length','Diameter','Height','Whole weight','Shucked weight'
                                 ,'Viscera weight','Shell weight'], y_vars='Rings',hue='Sex'
                                 , size=4, aspect=0.7)
```

# Step 5 : Preparing X and y

```python
#First of all we should create a list of feature names
features = ['Length','Diameter','Height','Whole weight','Shucked weight','Viscera weight',
            'Shell weight']

#Use these list as to select the subset of Series from pandas dataframe
X = abalone_df[features]
y = abalone_df['Rings']

#Print the dimensions of X and y
print "Dimensions of X : " + str(X.shape)
print "Dimensions of y : " + str(y.shape)
```

# Step 6 : Splitting X and y into training and testing sets

```python
from sklearn.cross_validation import train_test_split
#By default sklearn divides 75% training data and 25% testing data
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=0)
print "Dimensions of X_train : " + str(X_train.shape)
print "Dimensions of X_test : "  + str(X_test.shape)
print "Dimensions of y_train : " + str(y_train.shape)
print "Dimensions of y_test : "  + str(y_test.shape)
```

# Step 7 : Ridge regression using scikit-learn

```python
#Import Reige Regression Model
from sklearn.linear_model import Ridge

#Create a Ridge regression object
ridge_reg = Ridge()

#Train the model using the training set
#When alpha not set default is 1.0 for sklearn
ridge_reg.fit(X_train, y_train)
```

```python
#Print the learned intercept and coefficients
print "Intercept : " + str(ridge_reg.intercept_) 
zip(features, ridge_reg.coef_)
```

# Step 8 : Make predictions using model learned

```python
#We will make predictions using the testing set
y_pred = ridge_reg.predict(X_test)
```

# Step 9 : Evaluate the model using RMSE

```python
#Calculate the RSS
RSS = np.mean((y_pred - y_test)**2)
print "Ridge Regression RSS : " + str(RSS)

#Calculate the RMSE
RMSE = np.sqrt(RSS)
print "Ridge Regression RMSE : " + str(RMSE)
```

```python
#Variance score : 1 is perfect prediction
print "Variance score : " + str(ridge_reg.score(X_test,y_test))
```

# Step 10 : Let's try to see if considering only 3 parameters would have increased the performance
